# Relatório — Análise de Desempenho das Lojas Alura Store

## 1. Introdução

O objetivo deste relatório é auxiliar o **Senhor João** na decisão de **qual loja de sua rede Alura Store deve ser vendida**, possibilitando a obtenção de capital para investimento em um novo negócio.

Para isso, foi realizada uma análise exploratória dos dados de vendas das quatro lojas, considerando métricas financeiras, operacionais e de satisfação dos clientes. Além das análises numéricas, foram utilizados **gráficos gerados com a biblioteca Matplotlib**, facilitando a visualização dos padrões e apoiando a tomada de decisão baseada em dados.

---

## 2. Desenvolvimento das Análises

### 2.1 Faturamento Total das Lojas

A análise do faturamento total revelou diferenças claras entre as lojas. Conforme demonstrado no **gráfico de barras de faturamento**, a **Loja 1 apresentou o maior faturamento**, seguida pelas lojas 2 e 3.

A **Loja 4 apresentou o menor faturamento total**, indicando menor capacidade de geração de receita em comparação às demais lojas da rede. Esse fator é um dos principais indicadores de desempenho inferior.

---

### 2.2 Vendas por Categoria de Produto

Ao analisar a quantidade de vendas por categoria, observou-se que todas as lojas apresentam um **padrão semelhante de consumo**, com destaque para as categorias:

- Móveis  
- Eletrônicos  
- Brinquedos  

Apesar desse padrão comum, a **Loja 4 não se destaca positivamente em nenhuma categoria**, apresentando volumes de vendas consistentemente menores quando comparada às demais lojas. Isso indica uma **menor competitividade e menor demanda geral** por seus produtos.

Os gráficos de distribuição por categoria evidenciam que, mesmo com um mix semelhante, a Loja 4 apresenta desempenho inferior em termos de volume total.

---

### 2.3 Média de Avaliação dos Clientes

As avaliações dos clientes foram analisadas considerando uma escala consistente de **1 a 5**, confirmada pelos valores mínimos e máximos encontrados nos dados.

O **gráfico de linha das médias de avaliação** mostrou que:
- A **Loja 3 possui a melhor média de avaliação**, indicando maior satisfação dos clientes
- A **Loja 1 apresentou a menor média**
- A **Loja 4 possui uma avaliação intermediária**, sem destaque positivo em relação às demais

Assim, a Loja 4 não se sobressai em termos de experiência do cliente.

---

### 2.4 Produtos Mais e Menos Vendidos

A análise dos produtos mais e menos vendidos mostrou que as lojas possuem um portfólio semelhante. No entanto, existem diferenças importantes no comportamento das vendas.

A **Loja 4 apresenta grande variação no desempenho dos produtos**, dependendo de poucos itens com bom volume de vendas, como *Cama box* e *Faqueiro*. Em contrapartida, diversos produtos apresentam **baixo giro**, como *Guarda-roupas*, *Guitarra* e *Violão*.

Diferente das lojas 1, 2 e 3, que mantêm uma distribuição mais equilibrada de vendas entre os produtos, a Loja 4 demonstra **menor consistência operacional**, o que impacta negativamente seu desempenho geral.

---

### 2.5 Custo Médio de Frete por Loja

A análise do frete médio mostrou que a **Loja 4 possui o menor custo médio de frete** entre todas as lojas analisadas. Como o frete é pago pelo cliente, esse fator poderia representar uma vantagem competitiva.

Entretanto, ao cruzar essa informação com as demais métricas, observa-se que:
- O menor frete não resultou em maior faturamento
- Não impulsionou o volume de vendas
- Não gerou melhor avaliação dos clientes

Isso indica que o custo do frete, isoladamente, **não foi suficiente para melhorar o desempenho da Loja 4**.

---

## 3. Conclusão e Recomendação

A partir da análise conjunta dos dados e das visualizações gráficas geradas, conclui-se que a **Loja 4 apresenta o pior desempenho geral** entre as quatro lojas analisadas.

Mesmo possuindo o menor custo médio de frete, a Loja 4:
- Apresenta o **menor faturamento total**
- Não lidera vendas em nenhuma categoria de produto
- Não se destaca em satisfação dos clientes
- Possui grande número de produtos com baixo giro
- Demonstra menor consistência operacional

Dessa forma, recomenda-se que o **Senhor João venda a Loja 4**, pois essa decisão tende a causar **menor impacto financeiro e estratégico** na rede, além de liberar recursos para investimento em um novo negócio mais promissor.

---

## 4. Considerações Finais

Este estudo reforça a importância da **análise de dados aliada a visualizações gráficas** no apoio à tomada de decisões estratégicas. A avaliação integrada das métricas permitiu identificar, de forma clara e objetiva, a loja com menor desempenho, tornando a recomendação final mais segura e fundamentada.
